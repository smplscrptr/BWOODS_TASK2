```python
df.dropna(how="all", inplace=True)
df.rename(columns={"Unnamed: 0": "Year"}, inplace=True)
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")

df.sort_values(by="Year", inplace=True)
```

```python
df_interp = df.copy()
df_interp.set_index("Year", inplace=True)

df_interp = df_interp.interpolate(method="linear", axis=0)

df_interp.reset_index(inplace=True)

df_interp.head()
```

```python
df_long = df_interp.melt(id_vars="Year", var_name="Country", value_name="GDP")
df_long = df_long.dropna(subset=["GDP"])

df_long.head()
```

```python
latest_year = df_long["Year"].max()
top_countries = df_long[df_long["Year"] == latest_year].nlargest(5, "GDP")["Country"].tolist()

if "United States" not in top_countries:
    top_countries.append("United States")

df_top = df_long[df_long["Country"].isin(top_countries)]

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_top, x="Year", y="GDP", hue="Country")
plt.title("GDP Over Time for Selected Countries")
plt.ylabel("GDP")
plt.xlabel("Year")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```
